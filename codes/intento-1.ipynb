```python
def calculate_integral(A, B, C, D, k, lambda_, z, U0):
    # Define symbols
    x, y, x0, y0 = smp.symbols('x y x0 y0', real=True)

    # Define the integral expression
    integrand = (U0 * smp.exp(smp.I * k * A / (2 * B) * (x0**2 + y0**2)) *
                 smp.exp(-smp.I * k / B * (x * x0 + y * y0)))

    # Perform the double integral
    integral_result = smp.integrate(integrand, (x0, -smp.oo, smp.oo), (y0, -smp.oo, smp.oo))

    # Define the full expression
    U_xy = smp.I / (lambda_ * B) * smp.exp(smp.I * k * z) * smp.exp(smp.I * k * D / (2 * B) * (x**2 + y**2)) * integral_result

    return U_xy
```
